Solar potential of Pakistan based on average solar radiation (kwh/m^2/day)

Energy has an huge impact on economic prosperity of any country. From last 3 decades Pakistan facing a huge shortfall in its energy production and is now looking for alternate energy sources. Solar Energy is the best alternative source of energy as it also generates electricity. Also the solar energy is sustainable energy so its very good for environment.

Pakistan is a prime location for solar energy. It has extensive solar rays coming to it. It has certain regions that are extremely hot for over 6 months in a year. In fact, Pakistan has solar plants in certain cities. 

I have used the solar radiation data for all 12 months from 1977-2000 and averaged it for the whole year. I have visualized it for all the divisions of Pakistan and put the heavy transmissions lines on it. Idealy the solar radiation value of above 5 in kwh/m^2/day is good for installing solar panel.

To download datasets use the following sources/links
1) Raster dataset: https://biogeo.ucdavis.edu/data/worldclim/v2.1/base/wc2.1_30s_srad.zip 
2) Pakistan boundry shapefiles: https://biogeo.ucdavis.edu/data/gadm3.6/shp/gadm36_PAK_shp.zip 
3) Transmission lines and solar power plants location: https://www.re-explorer.org/re-data-explorer/data-library/layers 
4) Populated cities of Pakistan: https://simplemaps.com/data/pk-cities

```python
#The dataset has average solar radiation value for 12 months, I have to average all those raster layers to get an average for 1 year

def read_raster(name, as_type=float):
    r = rasterio.open(name)
    b = r.read(1)
    a = b.astype(as_type)
    return a

def average_rasters(rasters):

    # Empty list to append input raster values
    all_read_rasters = []

    total_input_rasters = len(rasters)

    for r in range(0, total_input_rasters):
        rr = read_raster(rasters[r])
        all_read_rasters.append(rr)

    array_pixel_values = np.array(all_read_rasters)

    mean_array = np.nanmean(array_pixel_values, axis=0)
    
    # Opening the first raster
    temp = rasterio.open(rasters[0])

    with rasterio.open('average_raster.tif', 'w', driver='Gtiff',
                       width=temp.width, height=temp.height,
                       count=1, crs=temp.crs, transform=temp.transform,
                       dtype='float64') as average_raster:
        average_raster.write(mean_array, 1)
        average_raster.close()

#importing the raster files to be taking mean
source_rasters = ['1.tif', '2.tif', '3.tif','4.tif','5.tif','6.tif','7.tif','8.tif','9.tif','10.tif','11.tif','12.tif']

#applying the function of averaging 
average_rasters(source_rasters)
```

```python
from rasterstats import zonal_stats
```

```python
#summarising the solar radiation per division with zonal mean

zonal_statistics = zonal_stats('gadm36_PAK_2.shp', 'average_raster.tif', stats=['mean'])
```

```python
print(zonal_statistics)
```

```python
#storing the mean into a geodataframe

zonalstats_df = pd.DataFrame(zonal_statistics)
```

```python
#renaming the column into mean_solar

zonalstats_df.rename(columns={'mean':'mean_solar'}, inplace=True)
```

```python
#combining the dataset of location and solarradiation together

pak_divisions = pd.concat([pak_divisions, zonalstats_df], axis=1)
```

```python
#visualizing it

pak_divisions.head(2)
```

```python
#ploting it

fig, ax = plt.subplots(figsize=(14,10))
plt.title("Average solar radiation (kJ/m^2/day) of Pakistan (1970-2000)", fontsize= 20)
pak_divisions.plot(column='mean_solar', ax=ax, legend=True, cmap= 'YlOrBr', edgecolor='black', lw=0.6)
plt.tight_layout()
```
